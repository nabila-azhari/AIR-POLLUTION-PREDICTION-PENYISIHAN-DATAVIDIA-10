```python
import pandas as pd

BASE = "submission_FINAL_SAFE.csv"
OUT  = "submission_FINAL_ALLIN.csv"

df = pd.read_csv(BASE)

if "kategori" in df.columns:
    df = df.rename(columns={"kategori": "category"})

df["tanggal"] = pd.to_datetime(df["id"].str.split("_").str[0])
df["lokasi"]  = df["id"].str.split("_").str[1]

# ===============================
# FINAL ALL-IN SHIFT
# ===============================

mask = (
    (df["tanggal"] >= "2025-11-24") &
    (df["tanggal"] <= "2025-11-30") &
    (df["lokasi"].isin(["DKI1","DKI2","DKI3","DKI5"]))
)

df.loc[mask, "category"] = "BAIK"

# ===============================
# SAVE
# ===============================

df = df[["id","category"]]
df.to_csv(OUT, index=False)

print("✅ FINAL ALL-IN saved:", OUT)
print("\nDistribusi akhir:")
print(df["category"].value_counts(normalize=True))
```
