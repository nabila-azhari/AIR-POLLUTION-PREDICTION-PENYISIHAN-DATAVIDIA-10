```python

import pandas as pd
from pathlib import Path

INPUT_DIR = Path(r"C:\Users\USER\Desktop\DATAVIDIA\penyisihan-datavidia-10\ispu_named\output")
FINAL_OUTPUT = Path(r"C:\Users\USER\Desktop\DATAVIDIA\penyisihan-datavidia-10\ispu_named\ispu_all_years.csv")
```

```python

dfs = []

for file in sorted(INPUT_DIR.glob("*.csv")):
    df = pd.read_csv(file)
    print(f"{file.name}: {df.shape[0]} rows")
    dfs.append(df)

df_all = pd.concat(dfs, ignore_index=True)
```

```python

print("Total rows setelah merge:", df_all.shape[0])
print("Total kolom:", df_all.shape[1])
display(df_all.head())
```

```python

df_all.to_csv(FINAL_OUTPUT, index=False)
```
