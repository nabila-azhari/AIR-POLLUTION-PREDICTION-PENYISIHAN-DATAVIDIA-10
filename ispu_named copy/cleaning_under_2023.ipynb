```python
from pathlib import Path
import pandas as pd
import re

BASE_DIR = Path.cwd()
PROJECT_ROOT = BASE_DIR.parent    # naik 1 level
ISPU_DIR = PROJECT_ROOT / "ISPU"    # ambil folder ISPU
OUTPUT_DIR = BASE_DIR / "input"       # ispu_named copy/input

OUTPUT_DIR.mkdir(exist_ok=True)


print("BASE_DIR      :", BASE_DIR)
print("PROJECT_ROOT  :", PROJECT_ROOT)
print("ISPU_DIR      :", ISPU_DIR)
```

```python
csv_files = sorted(ISPU_DIR.glob("*.csv"))

print("Total files:", len(csv_files))
for f in csv_files:
    print("-", f.name)
```

```python
def extract_year(path):
    match = re.search(r'20\d{2}', str(path))
    return int(match.group()) if match else 0
```

```python
files_sorted = sorted(csv_files, key=extract_year)
from IPython.display import display

for file in files_sorted:
    print("=" * 100)
    print(f"FILE: {file.name}")
    print("=" * 100)

    df = pd.read_csv(file)
    

    if "tanggal" in df.columns:
        df["tanggal"] = pd.to_datetime(df["tanggal"], errors="coerce")
        df = df.sort_values("tanggal").reset_index(drop=True)

    display(df.head())
```

```python
raw_data = {}

for file in csv_files:
    year = extract_year(file)
    if year is None:
        continue
    
    raw_data[year] = pd.read_csv(file)
```

```python
def preprocess_ispu(df, year):
    df = df.copy()

    # =====================
    # FIX COLUMN SHIFT ANOMALY (2020 data)
    # =====================
    # In some rows, columns are shifted: the numeric location digit appears
    # in `critical`, the pollutant appears in `categori`, and the category
    # appears in `lokasi_spku`. Detect when `critical` is numeric and fix.
    if year == 2020 and "critical" in df.columns and "categori" in df.columns and "lokasi_spku" in df.columns:
        critical_numeric = pd.to_numeric(df["critical"], errors="coerce").notna()

        if critical_numeric.any():
            # Save originals for clarity
            orig_critical = df.loc[critical_numeric, "critical"].astype(str)
            orig_categori = df.loc[critical_numeric, "categori"].astype(str)
            orig_lokasi = df.loc[critical_numeric, "lokasi_spku"].astype(str)

            # New values:
            # - critical should be the pollutant (currently in `categori`)
            # - categori should be the category (currently in `lokasi_spku`)
            # - lokasi_spku should be 'DKI' + the digit (currently in `critical`)
            df.loc[critical_numeric, "critical"] = orig_categori
            df.loc[critical_numeric, "categori"] = orig_lokasi
            df.loc[critical_numeric, "lokasi_spku"] = "DKI" + orig_critical

    # =====================
    # KODE LOKASI (SOURCE)
    # =====================
    if "kode_lokasi" not in df.columns:
        if "stasiun" in df.columns:
            df["kode_lokasi"] = df["stasiun"]
        elif "lokasi_spku" in df.columns:
            df["kode_lokasi"] = df["lokasi_spku"]
        else:
            df["kode_lokasi"] = None
    
    # =====================
    # HANDLE TANGGAL KHUSUS 2016 (mixed format)
    # =====================
    if year == 2016:
        tanggal_raw = df["tanggal"].astype(str)

        parsed = pd.Series(pd.NaT, index=df.index)

        # Case 1: format YYYY-MM-DD
        mask_ymd = tanggal_raw.str.match(r"\d{4}-\d{2}-\d{2}")
        parsed[mask_ymd] = pd.to_datetime(
            tanggal_raw[mask_ymd],
            format="%Y-%m-%d",
            errors="coerce"
        )

        # Case 2: format DD/MM/YYYY
        mask_dmy = tanggal_raw.str.match(r"\d{2}/\d{2}/\d{4}")
        parsed[mask_dmy] = pd.to_datetime(
            tanggal_raw[mask_dmy],
            format="%d/%m/%Y",
            errors="coerce",
            dayfirst=True
        )

        df["tanggal"] = parsed.dt.strftime("%Y-%m-%d")


    # =====================
    # HANDLE TANGGAL
    # =====================
    if year == 2022:
        tanggal_raw = df["tanggal"]
        mask_numeric = pd.to_numeric(tanggal_raw, errors="coerce").notna()

        tanggal_parsed = pd.Series(pd.NaT, index=df.index)

        tanggal_parsed[mask_numeric] = pd.to_datetime(
            pd.to_numeric(tanggal_raw[mask_numeric]),
            unit="D",
            origin="1899-12-30"
        )

        tanggal_parsed[~mask_numeric] = pd.to_datetime(
            tanggal_raw[~mask_numeric],
            format='%Y-%m-%d',
            errors="coerce"
        )

        df["tanggal"] = tanggal_parsed.dt.strftime("%Y-%m-%d")

    else:
        # Detect if format is M/D/YYYY (e.g., 9/1/2013)
        tanggal_raw = df["tanggal"].astype(str)
        has_slash = tanggal_raw.str.contains("/", regex=False).any()
        
        if has_slash:
            # Try parsing with M/D/YYYY format first
            parsed = pd.to_datetime(
                tanggal_raw,
                format='%m/%d/%Y',
                errors="coerce"
            )
            
            # For any that failed, try after replacing / with -
            failed_mask = parsed.isna()
            if failed_mask.any():
                tanggal_normalized = tanggal_raw.str.replace("/", "-", regex=False)
                
                # For 2018, use dayfirst (dd/mm/yyyy)
                if year == 2018:
                    parsed[failed_mask] = pd.to_datetime(
                        tanggal_normalized[failed_mask],
                        format='%d-%m-%Y',
                        errors="coerce",
                        dayfirst=True
                    )
                else:
                    parsed[failed_mask] = pd.to_datetime(
                        tanggal_normalized[failed_mask],
                        format='%Y-%m-%d',
                        errors="coerce"
                    )
            
            df["tanggal"] = parsed.dt.strftime("%Y-%m-%d")
        else:
            # No slash, proceed with standard parsing
            # normalisasi separator
            df["tanggal"] = tanggal_raw.str.replace("/", "-", regex=False)

            # kasus khusus 2018 (dd/mm/yyyy)
            if year == 2018:
                df["tanggal"] = pd.to_datetime(
                    df["tanggal"],
                    format='%d-%m-%Y',
                    errors="coerce",
                    dayfirst=True
                ).dt.strftime("%Y-%m-%d")
            else:
                df["tanggal"] = pd.to_datetime(
                    df["tanggal"],
                    format='%Y-%m-%d',
                    errors="coerce"
                ).dt.strftime("%Y-%m-%d")

    # =====================
    # KODE LOKASI CLEAN
    # =====================
    df["kode_lokasi_clean"] = (
        df["kode_lokasi"]
        .astype(str)
        .str.split(" ")
        .str[0]
    )

    # =====================
    # ID
    # =====================
    df["id"] = df["tanggal"] + "_" + df["kode_lokasi_clean"]

    return df
```

```python
processed_data = {}

for year, df in raw_data.items():
    processed_data[year] = preprocess_ispu(df, year)
    
```

```python
for year, df in processed_data.items():
    mask = df["tanggal"].notna()
    df.loc[mask, "tanggal"] = (
        df.loc[mask, "tanggal"]
        .astype(str)
        .str.slice_replace(0, 4, str(year))
    )
```

```python
for year in sorted(processed_data.keys()):
    print(f"DATA {year}")
    display(processed_data[year].head())
```

```python
for year in sorted(raw_data.keys()):
    print(
        f"{year} | raw: {len(raw_data[year])} | processed: {len(processed_data[year])}"
    )
```

```python
for file in csv_files:
    year = extract_year(file)

    if year is None or year > 2022:
        continue  # jangan sentuh 20222025

    print(f"Processing {year}  {file.name}")

    df_raw = pd.read_csv(file)
    df_processed = preprocess_ispu(df_raw, year)
    
    # =====================
    # FIX YEAR IN TANGGAL
    # =====================
    mask = df_processed["tanggal"].notna()
    df_processed.loc[mask, "tanggal"] = (
        df_processed.loc[mask, "tanggal"]
        .astype(str)
        .str.slice_replace(0, 4, str(year))
    )
    
    # =====================
    # FIX YEAR IN PERIODE_DATA
    # =====================
    if "periode_data" in df_processed.columns:
        mask_periode = df_processed["periode_data"].notna()
        df_processed.loc[mask_periode, "periode_data"] = (
            df_processed.loc[mask_periode, "periode_data"]
            .astype(str)
            .str.slice_replace(0, 4, str(year))
        )

    df_processed.to_csv(
        OUTPUT_DIR / f"{year}_id.csv",
        index=False
    )

    display(df_processed.head(2))
```
